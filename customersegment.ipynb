```python
import pandas as pd
```

```python


# Load Excel file
df = pd.read_excel("/content/drive/MyDrive/python practice/Online Retail.xlsx")

# Display first few rows
print(df.head())
```

```python
# Drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Ensure InvoiceDate is in datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create a TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Display first few rows
print(df.head())
```

```python
# Get the latest transaction date
latest_date = df['InvoiceDate'].max()

# Group by CustomerID to calculate RFM metrics
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency (Total transactions)
    'TotalPrice': 'sum'  # Monetary Value (Total spending)
})

# Rename columns
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Display results
print(rfm.head())
```

```python
# Save as CSV
rfm.to_csv('customer_segments.csv', index=False)
print("Customer segmentation data saved!")
```

```python
# Re-load the original dataset with CustomerID
df_original = pd.read_excel("/content/drive/MyDrive/python practice/Online Retail.xlsx")

# Drop missing CustomerID rows
df_original = df_original.dropna(subset=['CustomerID'])

# Convert CustomerID to integer type (to avoid decimal formatting)
df_original['CustomerID'] = df_original['CustomerID'].astype(int)

# Convert InvoiceDate to datetime
df_original['InvoiceDate'] = pd.to_datetime(df_original['InvoiceDate'])

# Create TotalPrice column
df_original['TotalPrice'] = df_original['Quantity'] * df_original['UnitPrice']

# Compute Recency, Frequency, and Monetary values while keeping CustomerID
latest_date = df_original['InvoiceDate'].max()

rfm = df_original.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'TotalPrice': 'sum'  # Monetary
}).reset_index()  # Keep CustomerID

# Rename columns
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Apply K-Means Clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Save the corrected dataset with CustomerID
save_path = "/content/drive/MyDrive/python practice/customer_segment_final1.csv"
rfm.to_csv(save_path, index=False)

print(f"✅ Final dataset with CustomerID saved successfully at: {save_path}")
```
